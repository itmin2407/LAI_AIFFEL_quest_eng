# 딕셔너리에 key value 넣기
---
### 1. `딕셔너리.setdefault(key, value)`로 초기값 넣기

- 이 메서드는 초기값만 넣을 수 있음

````python
# 빈 딕셔너리 'd' 생성
d = {}
d
````

````python
# 'a' 키가 없으면 'a':'에이'를 추가합니다. 이미 있다면 그대로 둡니다.
d.setdefault('a', '에이')
````

````python
# 확인
d
````

````python
# 'a' 키가 이미 존재하므로, setdefault 메서드는 딕셔너리를 변경하지 않습니다.
d.setdefault('a', '가')
````

````python
# # 확인 -> 이전과 동일함
d
````

### 2. `딕셔너리.update(key=value)`로 Key, Value를 여러번 넣기
- `.setdefault(key, value)`는 해당 키를 최초 한번만 만들어줄 수 있음.  
- `.update(key=value)`는 해당 키를 여러번 덮어쓸 수 있음.

````python
# 잘못된 사용 예시 -> setdefault와 사용법이 다름
d.update('a', '가')
````

````python
# 올바른 사용
d.update(a='가')
````

````python
# 확인
d
````

`d.update(a='가')` 괄호안의 a='가' 와 같이 데이터를 넣는 형식을 키워드 인자라고 함.

````python
d.update(b='비')
````

````python
d
````

### [의문] `.update()`를 사용할 때, 숫자를 키로 넣을 수는 없을까? (중요하진 않음)  
예를 들면 이렇게
```
d = {'a': '가', 'b': '비', 1: '일'}
```

````python
# 시도1
d.update(1='일')
````

````python
# 시도2
d.update('1'='일')
````

`.update(키=밸류)`  여기서 '키'자리에 들어간 값은 **파이썬 내부적으로(파서)** 변수명으로 취급함.  
파이썬 변수명 규칙에 따르면 숫자는 변수가 될 수 없음.  
그래서 `# 시도2`에서 본 것처럼 문법 에러가 발생함.  
  
그렇다면


````python
# 시도3 (M3의 삽질)

# 숫자 1을 먼저 다른 변수에 담고
z = 1

# 이렇게 하면 어떻게 될까?
d.update(z = '일')     # 우리의 기대 -> { 1: '일' }
````

````python
# 확인
d
````

`d.update(z = '일')` 이렇게 입력했는데 `{1: '일'}`이 되지 않고 `{'z': '일'}`이 됨.  
왜?   
방금 입력한 `(z=z)`는 키워드 인자 형태이기 때문.   
키워드 인자에서 '=' 부호를 기준으로 왼쪽은 **변수명**으로 만들고 오른쪽은 밸류값으로 넣게 됨.  

````python
z = 1     # <-이 값은 사용되지 않은 것
d.update(z = '일')     # { 1: '일'} 을 기대했지만...

# 결과 -> { 'z': '일'}
````

---

````python
# 이렇게 하면 어떻게 될까요?
z = 1
d.update(z = z)
````

데이터의 값 전달 방식¶
x['e'] = 50 (인덱싱/서브 스크립트 형식):

문자열(값)을 직접 사용하는 방식, ['e'] 안에 들어가는 것은 변수 이름(식별자) 아닌 데이터 자체로 자유도가 높음
x[100] = 50 (숫자 키 가능)
x["my-key!"] = 50 (특수문자/공백 가능)
x[('a', 'b')] = 50 (튜플도 가능)
x.setdefault('e', 50) (위치 인자 형식) 이것은 함수에 데이터를 전달할 때 사용하는 "위치 인자(Positional Arguments)" 형식 형식: 메서드(인자1, 인자2) 특징: update(a=10)에서는 a가 식별자였지만,

setdefault('e', 50)에서 'e'는 그냥 함수에 전달되는 첫 번째 데이터(값)일 뿐임 작동: 함수 내부에서 "첫 번째로 들어온 데이터('e')를 키로 쓰고, 두 번째 데이터(50)를 기본값으로 써야지"라고 정해진 순서대로 처리
x.update(key=vale) 형식에서 키는 식별자(identifier) 규칙을 따르는 키워드 인자(keyword Argmet) 문법 사용. 문법적 형식: 파이썬의 키워드 인자(Keyword Argument) 문법을 사용함. 식별자 사용: = 왼쪽에 오는 a는 변수명과 같은 식별자여야 함. (그래서 "a"=10이나 1=10처럼 쓸 수 없음) 데이터 변환: update 메서드가 실행되면서, 식별자로 들어온 a를 문자열 'a'로 변환하여 딕셔너리의 키(key)로 저장함. 문자 "10"도 식별자 규칙을 만족시키지 못함 즉, x.update('10'=50)도 x.update(10=50)가 되므로 사용할 수 없다.
